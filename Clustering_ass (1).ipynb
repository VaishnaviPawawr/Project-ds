# 1. Perform Clustering(Hierarchical, Kmeans & DBSCAN) for the crime data and identify the number of clusters formed and draw inferences.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.cluster.hierarchy as sch

from sklearn.cluster import AgglomerativeClustering
import warnings
warnings.filterwarnings('ignore')
```

Hierarchical 

```python
crime = pd.read_csv('crime_data.csv')
crime
```

```python
crime.info()
```

```python
crime.describe()
```

```python
def norm_fun(i):
    x = (i-i.min())/(i.max()-i.min())
    return x
```

```python
crime_norm = norm_fun(crime.iloc[:,1:])
crime_norm
```

```python
dendrogram = sch.dendrogram(sch.linkage(crime_norm, method='average'))
```

```python
hc = AgglomerativeClustering(n_clusters=3,affinity='euclidean',linkage='complete')
hc
```

```python
y_hc = hc.fit_predict(crime_norm)
y_hc
```

```python
crime['h_clusterid']=hc.labels_
crime
```

Kmeans

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.cluster import KMeans
from scipy.spatial.distance import cdist
```

```python
crime1 = pd.read_csv('crime_data.csv')
crime1
```

```python
def norm_func(i):
    x = (i-i.min())/(i.max()-i.min())
    return x
```

```python
crime1_norm=norm_func(crime1.iloc[:,1:])
```

```python
wcss = []
for i in range(1,11):
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(crime1_norm)
    wcss.append(kmeans.inertia_)
plt.plot(range(1,11),wcss)
plt.title('Elbow curv')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
model = KMeans(n_clusters=3)
model.fit(crime1_norm)
model.labels_
```

```python
x = pd.Series(model.labels_)
crime1['Clust'] = x
```

```python
crime1.iloc[:,1:5].groupby(crime1.Clust).mean()
```
